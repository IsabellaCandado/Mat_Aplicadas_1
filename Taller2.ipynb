###### <font color="1aa293"> **1.** </font> **Serie Fibonacci**
Escribe un programa en Python que genere los primeros $n$ números de la serie de Fibonacci y visualízalos en un gráfico de líneas utilizando Matplotlib. El
usuario debe poder elegir el número $n$.

```python
print("Sucesión de Fibonacci")
n=int(input("Ingrese el número de elementos de la sucesión: "))
print("Los primeros",n,"elementos de la sucesión de Fibonacci son:")
a=0
b=1
x=[]
y=[]
for i in range(n):
    x.append(i)
    c=a+b
    y.append(a)
    a=b
    b=c

print(y)
plt.plot(x,y,'g.',ms=10)
plt.grid(color='gray')
plt.axvline(0,color='black')
plt.axhline(0,color='black')
plt.title("Sucesión de Fibonacci")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.xlim(0,n)
plt.ylim(0,y[len(y)-1]+1)
plt.show()
```

###### <font color="1aa293"> **2.** </font> **Análisis de datos meteorológicos:**
Dado un conjunto de datos con las temperaturas diarias de una ciudad durante un año, escribe un programa que calcule la media, la mediana y la desviación estándar de las temperaturas. Luego, crea un gráfico de líneas que muestre la temperatura media mensual.


```python
temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4,26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9,
26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

media = sum(temperaturas) / len(temperaturas)
print("La media de las temperaturas es:", media)

mediana = np.median(temperaturas)
print("La mediana de las temperaturas es:", mediana)

desviacion_estandar = np.std(temperaturas)
print("La desviación estándar de las temperaturas es:", desviacion_estandar)

dias = range(1, len(temperaturas) + 1)
plt.plot(dias, temperaturas, marker='o', linestyle='-', color='blue')
plt.title('Temperatura Diaria del Mes')
plt.xlabel('Día del Mes')
plt.ylabel('Temperatura (°C)')
plt.xlim(1, len(temperaturas))
plt.grid("gray")
plt.show()
```

###### <font color="1aa293"> **4.** </font> **Gráfica de funciones trigonométricas:**
Escribe un programa que grafique las funciones seno, coseno y tangente en el mismo plano para valores entre $0$ y $2\pi$.


```python
x = np.linspace(0, 2*np.pi, 100)
y= [np.sin(x) for x in x]
z= [np.cos(x) for x in x]
t= [np.tan(x) for x in x]

plt.plot(x,y,"green") #Función seno
plt.plot(x,z,"orange") #Función coseno
plt.plot(x,t,'.') #Función tangente
plt.grid(color='gray')
plt.axvline(0,color='black')
plt.axhline(0,color='black')
plt.title("Funciones trigonométricas")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.xlim(0,2*np.pi)
plt.ylim(-1.5,1.5)
plt.show()
```

###### <font color="1aa293"> **6.** </font> **Pendiente de una recta:**
Escribe un programa que tome dos puntos en el plano $(x_1, y_1)$
y $(x_2, y_2)$, calcule la pendiente de la recta que los une, y grafique dicha recta.



```python
# @title Calculo de la pendiente y su grafica
x1 = 4 # @param {"type":"number","placeholder":"Ingrese su X1"}
y1 = 6 # @param {"type":"number","placeholder":"Ingrese su Y1"}
x2 = 2 # @param {"type":"number","placeholder":"Ingrese su X2"}
y2 = 7 # @param {"type":"number","placeholder":"Ingrese su Y2"}
import numpy as nu
import matplotlib.pyplot as plt
def calculoPendiente(x1,x2,y1,y2):
  if float(x2)-float(x1) == 0:
    raise ValueError ("La pendiente es indefinida")
  return (float(y2) - float(y1))/ (float(x2) - float(x1))

def calcularRecta (x1,x2,y1,y2,pendiente):
  xval = (float(x1), float(x2))
  yval = (float(y1), float(y2))

  plt.plot(xval, yval, marker='o', color='blue',)

  plt.grid("gray")
  plt.xlabel('Eje x')
  plt.ylabel('Eje y')
  plt.title('Gráfica de la recta')
  plt.axvline(0,color='black')
  plt.axhline(0,color='black')
  plt.show()

try:
    pendiente = calculoPendiente(x1, y1, x2, y2)

    print(f"La pendiente de la recta es: {pendiente}")

    calcularRecta(x1, y1, x2, y2, pendiente)
except ValueError as e:
    print(e)
```

###### <font color="1aa293"> **7.** </font> **Movimiento parabólico:**
Escribe un programa que simule el movimiento de un proyectil bajo la influencia de la gravedad (sin resistencia del aire). Grafica la trayectoria parabólica para diferentes ángulos de lanzamiento.

```python
print("Simula el lanzamiento de un proyectil según su angulo de inclinación:")
print("\nRecuerde que la velocidad de un proyectil oscila entre los 300 m/s y los 2000 m/s")
vel=int(input("Ingrese la velocidad del proyectil: "))
ang=int(input("Ingrese el ángulo de disparo: "))
g=9.8
x=[]
y=[]
b=True
for t in range(100):
    if b==True:
      x.append(vel*np.cos(np.radians(ang))*t)
      y.append(vel*np.sin(np.radians(ang))*t-0.5*g*(t**2))
      if y[t]<0:
        b=False
    else:
        break

plt.title("Movimiento del proyectil respecto al tiempo")
plt.xlabel("Tiempo")
plt.ylabel("Altura")
plt.grid("gray")
plt.plot(x,y,".")
plt.show()
```
